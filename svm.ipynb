```r
spamtib<-as_tibble(spam)
spamtib$type
#svm cannot handle categorical predictors solo continuous
```

```r
spamtask<-makeClassifTask(data = spamtib, target = 'type')
```

```r
makeLearner('classif.svm')
```

```r
getParamSet('classif.svm')

#it will scale the data for us!
```

```r
getParamSet('classif.svm')$pars$kernel$values
```

```r
kernels<-c('polynomial','radial','sigmoid')

svmparamspace<-makeParamSet(
    makeDiscreteParam('kernel',values = kernels),
    makeIntegerParam('degree', lower = 1, upper = 3),
    makeNumericParam('cost', lower=0.1, upper = 10),
    makeNumericParam('gamma', lower=0.1, upper = 10)
)
```

```r
svmparamspace
```

```r
100*100*9
```

```r
randsearch<-makeTuneControlRandom(maxit = 500)
```

```r
#holdout
cvfortuning<-makeResampleDesc('Holdout', split = 2/3)
```

```r
parallel::detectCores()
```

```r
library(parallelMap)
library(parallel)
```

```r
parallelStartSocket(cpus = detectCores())

tunedsvmpars<-tuneParams('classif.svm', task = spamtask, resampling = cvfortuning, control = randsearch, par.set = svmparamspace)

parallelStop()
```

```r
#Op. pars: kernel=polynomial; degree=1; cost=6.97; gamma=6.18
bestparams<-tunedsvmpars$x
```

```r
saveRDS(bestparams, 'best_hyperparameters.rds')
```

```r
bestparams<-readRDS('best_hyperparameters.rds')
bestparams
```

```r
setHyperPars(makeLearner('classif.svm'), par.vals = bestparams)
```
